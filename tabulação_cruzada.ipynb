```python
import pandas as pd

network = [
    ("Allan", "Elias"),
    ("Allan", "Pierry"),
    ("Luana", "Tavares"),
    ("Luana", "pierry"),
    ("Pierry", "Elias"),

]

# created dataframe of the network
friends1 = pd.DataFrame(
    network, columns=["Person1", "Person2"]
)

#reverse the order of lolumns
friends2 =pd.DataFrame(
    network, columns=["Person2", "Person1"]
)

#created symmetric dataframe
friends = pd.concat([friends1, friends2])

# created a cross tubulation
pd.crosstab(friends.Person1, friends.Person2)
```
